Logistic Regression Model to Classify Rock and Mines Under The Sea Level

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Processing

```python
#Loading Dataset into pandas Dataframe
```

```python
sonar_data = pd.read_csv('/content/Copy of sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
#check rows and coloumns
sonar_data.shape
```

```python
sonar_data.describe()  #to describe statistical measures for data
```

```python
sonar_data[60].value_counts()
```

This is a small dataset, to create efficient ML models we need tons of data.
Here M= Mines and R= Rocks
Since the values of M & R are almost equal we can give much accurate predictions

```python
sonar_data.groupby(60).mean()
```

```python
#Now we need to seperate the Data and Labels(M, R)
X= sonar_data.drop(columns= 60, axis=1)
Y= sonar_data[60]
```

```python
print(X)
```

```python
print(X)
print(Y)
```

```python
#Splitting this data again into "Training" and "Test" Data
```

**Training and Test Data**

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)

#test_size 0.1 states that 10% of our data is testing data
#stratify is to split data equally in rock and mines
#random state to reporduce code in same manner uniformly
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training Step: Logistic Regression

```python
model= LogisticRegression()
```

```python
#Training Logisitc Regression Model with Training Data

model.fit(X_train, Y_train)
```

"Model Evaluation"

Training Data Accuracy Check:

```python
X_train_prediction= model.predict(X_train)
training_data_acc= accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training Data:", training_data_acc)
```

Test Data Accuracy Check:

```python
X_test_prediction= model.predict(X_test)
test_data_acc= accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on Test Data:", test_data_acc)
```

Accuracy Score > 70%

Thus, our model is predicting correctly
